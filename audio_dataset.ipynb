```python
#importing the libraries

from os import listdir
import os
import numpy as np
from PIL import Image
import shutil
import pandas as pd
import urllib.request
import IPython
import matplotlib.pyplot as plt
from prettytable import PrettyTable
```

```python
# function to plot bar graph

def plot_bar_graph(label, no_images, title):
    index = np.arange(len(label))
    plt.bar(index, no_images)
    plt.xlabel('Emotions')
    plt.ylabel('No of Images')
    plt.xticks(index, label, rotation=30)
    plt.title(title)
    plt.show()
```

# Savee Dataset
http://kahlan.eps.surrey.ac.uk/savee/

Surrey Audio-Visual Expressed Emotion (SAVEE) database has been recorded as a pre-requisite for the development of an automatic emotion recognition system. The database consists of recordings from 4 male actors in 7 different emotions, 480 British English utterances in total.<br>
Speakers 'DC', 'JE', 'JK' and 'KL' are four male speakers recorded for the SAVEE database

Audio files consist of audio WAV files sampled at 44.1 kHz

There are 15 sentences for each of the 7 emotion categories.
The initial letter(s) of the file name represents the emotion class, and the following digits represent the sentence number.
The letters 'a', 'd', 'f', 'h', 'n', 'sa' and 'su' represent 'anger', 'disgust', 'fear', 'happiness', 'neutral', 'sadness' and 'surprise' emotion classes respectively. 
E.g., 'd03.wav' is the 3rd disgust sentence. 




```python
base_path_audio = 'D:\\Aakash\\Documents\\temp\\emotion detection dataset\\audio expression\\savee_1\\AudioData'
final_data_path = 'D:\\Aakash\Documents\\temp\\emotion detection dataset\\final audio dataset\\savee'
```

```python
# iterating through all the folders and make list of their paths and renaming the audio files

folders = listdir(base_path_audio)
for folder in folders:
    audios = listdir(base_path_audio + '\\' + folder)
    for audio in audios:
        os.rename(base_path_audio + '\\' + folder + '\\' + audio, base_path_audio + '\\' + folder + '\\' + audio[:-4] + '_' + folder + '.wav')
```

```python
# iterating through all the audio path and saving in the corresponding emotion folder

folders = listdir(base_path_audio)
for folder in folders:
    audios = listdir(base_path_audio + '\\' + folder)
    for audio in audios:
        x = audio[0]
        if x == 'n':
            shutil.move(base_path_audio + '\\' + folder + '\\' + audio, final_data_path + '\\neutral')
        if x == 'h':
            shutil.move(base_path_audio + '\\' + folder + '\\' + audio, final_data_path + '\\happiness')
        if x == 's':
            if audio[1] == 'a':
                shutil.move(base_path_audio + '\\' + folder + '\\' + audio, final_data_path + '\\sadness')
            else:
                shutil.move(base_path_audio + '\\' + folder + '\\' + audio, final_data_path + '\\surprise')
        if x == 'a':
            shutil.move(base_path_audio + '\\' + folder + '\\' + audio, final_data_path + '\\anger')
        if x == 'd':
            shutil.move(base_path_audio + '\\' + folder + '\\' + audio, final_data_path + '\\disgust')
        if x == 'f':
            shutil.move(base_path_audio + '\\' + folder + '\\' + audio, final_data_path + '\\fear')
```

```python
#plotting the bar graph on emotion vs number of images

labels = listdir(final_data_path)
no_audio_savee = [len(listdir(final_data_path+'\\'+label)) for label in labels]

plot_bar_graph(labels, no_audio_savee, 'Bar plot for SAVEE dataset')
```

```python
pt = PrettyTable()
pt.field_names = ['Emotion', 'No. of images']
pt.add_row(['Anger', no_audio_savee[0]])
pt.add_row(['Disgust', no_audio_savee[1]])
pt.add_row(['Fear', no_audio_savee[2]])
pt.add_row(['Happiness', no_audio_savee[3]])
pt.add_row(['Neutral', no_audio_savee[4]])
pt.add_row(['Sadness', no_audio_savee[5]])
pt.add_row(['Surprise', no_audio_savee[6]])

print(pt)
```

# RAVDESS Dataset
https://zenodo.org/record/1188976#.XZERqEYzbIU

The Ryerson Audio-Visual Database of Emotional Speech and Song (RAVDESS) contains 7356 files (total size: 24.8 GB). The database contains 24 professional actors (12 female, 12 male), vocalizing two lexically-matched statements in a neutral North American accent. Speech includes calm, happy, sad, angry, fearful, surprise, and disgust expressions, and song contains calm, happy, sad, angry, and fearful emotions. Each expression is produced at two levels of emotional intensity (normal, strong), with an additional neutral expression. All conditions are available in three modality formats: Audio-only (16bit, 48kHz .wav), Audio-Video (720p H.264, AAC 48kHz, .mp4), and Video-only (no sound).  Note, there are no song files for Actor_18.

Audio-only files of all actors (01-24) are available as two separate zip files (~200 MB each):

Speech file (Audio_Speech_Actors_01-24.zip, 215 MB) contains 1440 files: 60 trials per actor x 24 actors = 1440. 
Song file (Audio_Song_Actors_01-24.zip, 198 MB) contains 1012 files: 44 trials per actor x 23 actors = 1012.

# TESS Dataset
https://tspace.library.utoronto.ca/handle/1807/24487

Toronto emotional speech set (TESS) Collection: 
A set of 200 target words were spoken in the carrier phrase "Say the word _____' by two actresses (aged 26 and 64 years) and recordings were made of the set portraying each of seven emotions (anger, disgust, fear, happiness, pleasant surprise, sadness, and neutral). There are 2800 stimuli in total.
Two actresses were recruited from the Toronto area. Both actresses speak English as their first language, are university educated, and have musical training. Audiometric testing indicated that both actresses have thresholds within the normal range.

```python
base_url = 'https://tspace.library.utoronto.ca/bitstream/1807/'
final_data_path = 'D:\\Aakash\\Documents\\temp\\emotion detection dataset\\final audio dataset\\Torronto_speech_dataset'
base_path_files = 'D:\\Aakash\\Documents\\temp\\emotion detection dataset\\final audio dataset\\Torronto_speech_dataset\\happiness'
lst_files = listdir(base_path_files)
lst_words = [word.split('_')[1] for word in lst_files]
lst_words.sort(reverse = True)
```

```python
def download_files(emotion, age, location, num):
    i = 1
    for word in lst_words:
        try:
            urllib.request.urlretrieve(base_url+str(num)+'/'+str(i)+'/'+age+'_'+ word+'_'+emotion+'.wav', location+word+age+str(i)+'.wav')
            i += 1
        except IOError:
            print(word)
```

```python
download_files('happy', 'YAF', base_path+'\\happiness\\', 24493)
download_files('sad', 'YAF', base_path+'\\sadness\\', 24500)
download_files('sad', 'OAF', base_path+'\\sadness\\', 24497)
download_files('angry', 'OAF', base_path+'\\anger\\', 24499)
download_files('angry', 'YAF', base_path+'\\anger\\', 24490)
download_files('fear', 'OAF', base_path+'\\fear\\', 24492)
download_files('fear', 'YAF', base_path+'\\fear\\', 24489)
download_files('neutral', 'OAF', base_path+'\\neutral\\', 24488)
download_files('neutral', 'YAF', base_path+'\\neutral\\', 24496)
download_files('disgust', 'YAF', base_path+'\\disgust\\', 24498)
download_files('disgust', 'OAF', base_path+'\\disgust\\', 24494)
download_files('ps', 'OAF', base_path+'\\surprise\\', 24491)
download_files('ps', 'YAF', base_path+'\\surprise\\', 24495)
```
